#### Missing values

```python
# CALCULATE ALL MISSING VALUES AND PLOT THEM IN A DATAFRAME IN DESCENDING ORDER

# Calculate the number of missing values for each column
missing_values_df = df.select([count(when(isnull(c), c)).alias(c) for c in df.columns])

# Calculate the percentage of corresponding null values
percents_df = df.select(
    [(count(when(isnull(c), c)) / lit(df.count()) * 100).alias(c)
     for c in df.columns])

# Collect the results into Python lists
missing_values = missing_values_df.collect()[0].asDict()
percents = percents_df.collect()[0].asDict()

missing_values_to_percents = {col: (missing_values[col], percents[col]) for col in missing_values}

# transform the dictionary into a list of tuples.
data_for_df = [(col, *vals) for col, vals in missing_values_to_percents.items()]

# Create a Pandas DataFrame from the data
df_as_dict = pd.DataFrame(data_for_df, columns=['Feature', 'Total Nulls', 'Percentage Nulls'])

# Display the dataframe, which honestly looks nicer to read
df_as_dict.sort_values(ascending=False, by='Percentage Nulls')
```

#### Variable transformations to plot histograms and visualize correlations

```python
# standardizes specific columns in a dataset
columns_to_standardize = ['depth', 'mag', 'nst', 'gap', 'dmin', 'rms', 'magNst']

# Sampling Data
percentage_to_sample = 0.20

# Remove Rows with Missing Values & Create a Sample:
df_no_nan = df.dropna(subset=columns_to_standardize)
df_sample = df_no_nan.sample(False, percentage_to_sample)

# Set Up the Transformation Pipeline:
stages = []
for col in columns_to_standardize:
    assembler = VectorAssembler(inputCols=[col], outputCol=col + "_vec")
    scaler = StandardScaler(inputCol=col + "_vec", outputCol=col + "_standardized")
    stages += [assembler, scaler]

# Creating and Apply the Pipeline:
pipeline = Pipeline(stages=stages)
model = pipeline.fit(df_sample)
df_transformed = model.transform(df_sample)

# Select and Convert Standardized Data to Pandas DataFrame:
scaled_columns = [col + "_standardized" for col in columns_to_standardize]
pandas_df = df_transformed.select(*scaled_columns).toPandas()
```

### Correlations between numerical variables

```python
# List of Columns to Correlate:
columns_to_correlate = ['latitude', 'longitude', 'depth', 'mag', 'nst', 'gap', 'dmin',
                        'rms', 'horizontalError', 'depthError', 'magError', 'magNst']

# Remove Rows with Null Values:
df_no_null = df.dropna(subset=columns_to_correlate)

# Select Only the Numerical Columns:
df_numeric = df_no_null.select(*columns_to_correlate)

# Prepare Vector Assembler:
assembler = VectorAssembler(inputCols=columns_to_correlate, outputCol="features")
df_vector = assembler.transform(df_numeric).select("features")

# Computing the Correlation Matrix:
correlation_matrix = Correlation.corr(df_vector, "features").head()[0].toArray()

# Converting to Pandas DataFrame:
pd_corr_matrix = pd.DataFrame(correlation_matrix, index=columns_to_correlate, columns=columns_to_correlate)

# Print the Correlation Matrix Dimensions:
print("Dimensions of the correlation matrix:", pd_corr_matrix.shape)

pd_corr_matrix
```

#### Imputing missing values for 'dmin' and 'gap' using the mean

```python
# Import the Imputer class from PySpark's ml.feature module for handling missing data
from pyspark.ml.feature import Imputer

# Initialize the Imputer with the input columns ('dmin' and 'gap') and output columns ('dmin' and 'gap') that will store the imputed values
# We are using the "mean" strategy to fill in the missing values
imputer = Imputer(
    inputCols=["dmin", "gap"],  # Columns with missing values to be imputed
    outputCols=["dmin", "gap"]  # Columns where the imputed values will be stored (same as input columns)
).setStrategy("mean")  # Strategy for imputing values (mean of the column in this case)

# Fit the imputer model on the data and apply the transformation
df = imputer.fit(df).transform(df)  # Apply the imputation transformation to the dataframe
```

### Imputation of the 'nst' column using LINEAR REGRESSION.

```python
# List of columns to use for feature engineering (numerical features for regression)
columns_to_use = ['latitude', 'longitude', 'depth', 'mag', 'rms']

# VectorAssembler: combines multiple columns into a single vector column 'features' for machine learning models
assembler = VectorAssembler(inputCols=columns_to_use, outputCol="features")

# StandardScaler: scales the 'features' column to standardize data (zero mean, unit variance) for better model performance
scaler = StandardScaler(inputCol="features", outputCol="scaledFeatures")

# Target variable: 'nst' (number of stations), which we want to predict
target = 'nst'

# Linear Regression: model to predict the target variable using the scaled features
lr = LinearRegression(featuresCol="scaledFeatures", labelCol=target)

# Pipeline: combines the assembler, scaler, and linear regression into a single workflow for training
pipeline = Pipeline(stages=[assembler, scaler, lr])

# Filter out rows with null values in the target column ('nst') for training
train_df = df.filter(df.nst.isNotNull())

# Fit the pipeline model on the training data
model = pipeline.fit(train_df)
full_predictions = model.transform(df)

df_imputed = full_predictions.withColumn(
    "nst_imputed",
    when(full_predictions["nst"].isNull(), full_predictions["prediction"])
    .otherwise(full_predictions["nst"])
)

# Drop unnecessary columns ('features', 'scaledFeatures', 'prediction') from the dataframe
df_final = df_imputed.drop("features", "scaledFeatures", "prediction")

df_final = df_final.drop("nst")

# Rename the imputed target column ('nst_imputed') back to the original column name ('nst')
df_final = df_final.withColumnRenamed("nst_imputed", "nst")
```

### Imputation of the 'magNst' column using a Decision Tree Regressor

```python
target = 'magNst'
columns_to_use = ['latitude', 'longitude', 'depth', 'mag', 'rms']

assembler = VectorAssembler(inputCols=columns_to_use, outputCol="features")


dt = DecisionTreeRegressor(featuresCol="features", labelCol=target)
pipeline = Pipeline(stages=[assembler, dt])

train_df = df_final.filter(df_final.magNst.isNotNull())
model = pipeline.fit(train_df)

full_predictions = model.transform(df_final)
df_imputed = full_predictions.withColumn(
    "magNst_imputed",
    when(full_predictions["magNst"].isNull(), full_predictions["prediction"])
    .otherwise(full_predictions["magNst"])
)

df_final = df_imputed.drop("features", "prediction")

df_final = df_final.drop("magNst").withColumnRenamed("magNst_imputed", "magNst")
```
